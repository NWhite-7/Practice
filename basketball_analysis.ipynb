```python
import pandas as pd
from matplotlib import pyplot as plt
import numpy as np
```

## Data preparation

```python
data = pd.read_csv('athlete_events.csv')
data
```

This is a pretty straightforward file, which makes it easy for us! 

We may be able to find a relationship between the height of basketball players and the number of gold medals they won. Let's see.

```python
basketball_data = data[data.Sport == 'Basketball']
basketball_data
```

4,536 rows is a lot! That may be enough data. Let's group these results to see the average height of players and the results for each team each year. We'll remove any entries that don't have heights for their players. 

```python
basketball_grouped = basketball_data.groupby(['Team', 'Year']).Height.mean().reset_index()
basketball_grouped.dropna(inplace = True)
basketball_grouped
```

Now let's add in whether they received a medal.

```python
medal = basketball_data[['Team', 'Year', 'Medal']]
medal_grouped = medal.groupby(['Team', 'Year']).Medal.last().reset_index()
medal_grouped
```

```python
basketball = pd.merge(basketball_grouped, medal_grouped, on = ['Team', 'Year'])
basketball
```

Beautiful. Now we have the team, the year they competed, the average height of their players, and which medal they received, if any. 

Let's plot this quickly to get a feel for the data. We'll need to convert the strings for the medals into numbers - we'll assign Gold a value of 3, Silver a value of 2, Bronze a value of 1, and no medal a value of 0.

```python
def medal_convert(x):
    if x == 'Gold':
        return 3
    elif x == 'Silver':
        return 2
    elif x == 'Bronze':
        return 1
    else:
        return 0

basketball['Medal_num'] = basketball.Medal.apply(medal_convert)
basketball
```

```python
plt.plot(basketball.Height, basketball.Medal_num, 'o')
plt.show()
```
